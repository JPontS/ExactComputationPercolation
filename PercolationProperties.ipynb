Import python modules and the edges of the network for which we are going to compute the percolation properties.

```python
fn_net = 'double_triangle.txt'
path_net = os.path.join('/Users/joanpont/Desktop/Master/MasterThesis/Networks/', fn_net)
edges = np.sort(np.loadtxt(path_net,int), axis=1)
```

We initialize the CCDs of all the edges:

```python
ls_CCDs = [CCD(nodes=edge, ls_roots=[(np.min(edge),np.min(edge)), 
                                    (np.min(edge),np.max(edge))]) for edge in edges]
```

We define an elimination order and we delete the nodes for which we are computing $p_{ij}$, e.g. $i=1$ and $j=6$.

```python
i,j = 1,6
ls_elimination = [5,6,4,3,2,1]
ls_elimination = [node for node in ls_elimination if node not in [1,6]]
```

```python
for node in ls_elimination:
    CCDs_node = [x for x in ls_CCDs if node in x._nodes] 
    reduce((lambda x, y: x.multiply(y)), CCDs_node)
    CCDs_node[0].remove_node(node)
    for x in CCDs_node[1:]:
        x.delete_CCD() 
```
